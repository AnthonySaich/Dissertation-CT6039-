```python
#This imports the dataset needed 
df = pd.read_csv("CleanedHAMandSPAMDataset.csv")
```

```python
#This marks 0 has HAM emails and 1 as SPAM emails
ham=df[df['spam']==0]
spam=df[df['spam']==1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
test_countVectorizer=vectorizer.fit_transform(df['text'].values.astype('U'))
```

```python
print(vectorizer.get_feature_names())
#
```

```python
test_countVectorizer.shape
```

```python
label=df['spam']
X=test_countVectorizer
y=label
```

```python
X.shape
#gets the shape of the X collom
```

```python
y.shape
#gets the shape of the Y collom
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
#inports train_test_split library 
```

```python
from sklearn.naive_bayes import MultinomialNB
#inports the naive_bayes library  
```

```python
NB_classifier=MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
#inports the classification_report,confusion_matrix library  
```

```python
y_predict_train=NB_classifier.predict(X_train)
y_predict_train

#outputs the y predict train
```

```python
cm=confusion_matrix(y_train,y_predict_train)
#binds the cm variable to y train and y predict train
```

```python
y_predict_test=NB_classifier.predict(X_test)
y_predict_test
```

```python
print(classification_report(y_test,y_predict_test))
#this prints the reports 
```
